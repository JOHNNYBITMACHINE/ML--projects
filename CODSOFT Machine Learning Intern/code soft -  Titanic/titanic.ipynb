```python
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, RobustScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score, confusion_matrix, roc_curve , roc_auc_score, accuracy_score

from sklearn import svm
```

```python
ord = OrdinalEncoder()
rob = RobustScaler()
sc = StandardScaler()
minn = MinMaxScaler()
```

```python
x = final.drop(columns='Survived')
y = final['Survived']
```

```python
x_train, x_test , y_train, y_test = train_test_split(x,y , test_size=0.2, random_state=24)
```

```python
x_train[['Fare','Pclass', 'Age', 'Embarked', 'family']] = sc.fit_transform(x_train[['Fare','Pclass', 'Age', 'Embarked', 'family']])
x_test[['Fare','Pclass', 'Age', 'Embarked', 'family']] = sc.fit_transform(x_test[['Fare','Pclass', 'Age', 'Embarked', 'family']])
```

```python
lor = LogisticRegression()
```

```python
lor.fit(x_train, y_train)
```

```python
lor.score(x_test, y_test)
```

```python
y_pred = lor.predict(x_test)
```

```python
def model_eval(X_test,y_test):
  y_pred=lor.predict(X_test)
  tn,fp,fn,tp=confusion_matrix(y_test,y_pred).ravel()
  accuracy=(tp+tn)/(tp+tn+fp+fn)
  precision=tp/(tp+fp)
  recall=tp/(tp+fn)
  specifivity=tn/(fp+tn)
  F1=2*(recall*precision)/(recall+precision)
  result={'accuracy':accuracy,'precision':precision,'recall':recall,'specifivity':specifivity,'F1':F1}
  return result
model_eval(x_test,y_test)
```
